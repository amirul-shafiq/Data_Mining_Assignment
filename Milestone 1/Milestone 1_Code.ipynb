# 1. Data Collection


Perform following operations:

__a) Import necessary library<br>__

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import matplotlib.pyplot as plt
import seaborn as sns
```

__b) Manually record company stock ticker code and create a list of URL of data info for each company <br>__

```python
# Telekom Malaysia : 4863
# Maxis : 6012
# Axiata : 6888
# Digi : 6947

telco_code = ['4863', '6012', '6888', '6947']
url =[]

for i in telco_code:
    website_url = ('https://www.klsescreener.com/v2/stocks/view/'+i)
    url.append(website_url)

url
```

__c) Scrape the data from each URL<br>__

```python
#get the data and append in data format
frames = []
for link in url:
    reso = requests.get(link)
    soup = BeautifulSoup(reso.text,'lxml')
    table = soup.find('table', {'class':'financial_reports table table-hover table-sm table-theme'})
    df = pd.read_html(str(table), header=0)
    df[0].rename(index= str, inplace = True)
    frames.append(df[0].assign(ticker=link[-4:]))
        
frames
```

__d) Compile data into a single data frame<br>__

```python
df = pd.concat(frames)
df
df.to_csv('df.csv')
```
